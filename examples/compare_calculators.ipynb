```python
import matplotlib.pyplot as plt
import numpy as np
import pyrvt

%matplotlib inline
```

```python
names = ['CLH56', 'D64', 'V75', 'DK85', 'LP99', 'Sea24']
```

```python
osc_freqs = np.geomspace(0.1, 100, num=91)
osc_damp = 0.05
```

```python
fig, ax = plt.subplots(figsize=(8,6))

for name in names:
    mot = pyrvt.motions.SourceTheoryMotion(magnitude=6.5, distance=20, region='wna', peak_calculator=name)
    mot.calc_fourier_amps()
    
    spec_accels = mot.calc_osc_accels(osc_freqs, osc_damp)

    ax.plot(osc_freqs, spec_accels, label=name)

ax.legend()
ax.set(xlabel='Osc. Freq.', xscale='log', ylabel='5%-Damped, Spec. Accel. (g)', yscale='log')
```
